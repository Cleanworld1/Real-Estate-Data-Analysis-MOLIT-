```python
import pandas as pd


df = pd.read_csv('bjdcode_20220313.csv', encoding='utf-8')
df
```

```python
df = df[df['폐지여부']=='존재']
df["법정동코드"]=df["법정동코드"].astype(str)
df["법정동코드"] = df.apply(lambda row: row["법정동코드"][0:5], axis=1)
df
```

```python
df2 = df.drop_duplicates(subset=["법정동코드"], keep="first")
```

```python
df2
```

```python
df2.to_csv('temp.csv', index=False, encoding="utf-8-sig")
```

```python
import csv

dict_from_csv = {}

with open('temp.csv', mode='r', encoding='utf-8') as inp:
    reader = csv.reader(inp)
    dict_from_csv = {rows[1]:rows[0] for rows in reader}

print(dict_from_csv)
```
